# Question Answering
## This is the roberta-base model, fine-tuned on the SQuAD2.0 dataset for Question Answering, including handling unanswerable questions.
### link hugging face https://huggingface.co/deepset/roberta-base-squad2


```python
import transformers
from transformers import pipeline, set_seed
```

```python
qa = pipeline('question-answering', model='deepset/roberta-base-squad2')
set_seed(42)
```

```python

QA_input = {
    'question': '''what's the importance of RAG in context Hallucination?''',
    'context': '''Retrieval-Augmented Generation (RAG) is a technique designed to reduce hallucinations in large language models (LLMs). Hallucinations occur when models generate content that appears plausible but is factually
                incorrect or unsupported by the data they were trained on.'''
}
res = qa(QA_input,
         top_k=3)
```

```python
for i in range(len(res)):
    print(f"Answer: '{res[i]['answer']}', score: {round(res[i]['score'], 4)}, start: {res[i]['start']}, end: {res[i]['end']}")
```
